```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Энергообъединение ОЭС': ['Центра', 'Средней Волги', 'Урала', 'Северо-Запада', 'Юга', 'Сибири'],
        'ТЭС МВт': [36070.23, 16203.48, 49979.59, 15572.14, 13757.29, 26577.96],
        'ГЭС МВт': [1800.07, 7013.00, 1901.19, 2947.24, 6289.69, 25301.60],
        'АЭС МВт': [14778.28, 4072.00, 1485.00, 5947.63, 4030.27, 0]}

df = pd.DataFrame(data)
df = df.sort_values(by=['ТЭС МВт', 'АЭС МВт'], ascending=[True, True]) # добавляем сортировку значений
df_numeric = df.select_dtypes(include='number') 
df['Всего'] = df_numeric.sum(axis=1) # добавляем новый столбец с суммой всех значений
df_new = df.loc[[1, 3], ['Энергообъединение ОЭС', 'Всего']]


print(df) # Вывод основной таблицы

print(df_new) # Вывод таблицы с двумя колонками и двумя строчками


# добавление гистограммы
plt.bar(df['Энергообъединение ОЭС'], df['ТЭС МВт'], color='y', alpha=0.5)
plt.bar(df['Энергообъединение ОЭС'], df['ГЭС МВт'], color='g', alpha=0.5)
plt.bar(df['Энергообъединение ОЭС'], df['АЭС МВт'], color='b', alpha=0.5)
plt.legend(['ТЭС МВт', 'ГЭС МВт', 'АЭС МВт'])
plt.xlabel('Энергообъединение ОЭС')
plt.ylabel('Мощность, МВт')
plt.title('Мощность энергообъединений ОЭС')
plt.show()




# добавление линейного графика
data = [36070.23, 16203.48, 49979.59, 15572.14, 13757.29, 26577.96]
labels = ['Центра', 'Средней Волги', 'Урала', 'Северо-Запада', 'Юга', 'Сибири']

plt.pie(data, labels=labels, autopct='%1.1f%%')
plt.title('Мощность энергообъединений ОЭС')
plt.show()



# добавление круговой диаграммы
data = {'Энергообъединение ОЭС': ['Центра', 'Средней Волги', 'Урала', 'Северо-Запада', 'Юга', 'Сибири'],
        'ТЭС МВт': [36070.23, 16203.48, 49979.59, 15572.14, 13757.29, 26577.96],
        'ГЭС МВт': [1800.07, 7013.00, 1901.19, 2947.24, 6289.69, 25301.60],
        'АЭС МВт': [14778.28, 4072.00, 1485.00, 5947.63, 4030.27, 0]}

x = data['Энергообъединение ОЭС']
y1 = data['ТЭС МВт']
y2 = data['ГЭС МВт']
y3 = data['АЭС МВт']

plt.plot(x, y1, color='r', marker='o')
plt.plot(x, y2, color='g', marker='o')
plt.plot(x, y3, color='b', marker='o')
plt.legend(['ТЭС МВт', 'ГЭС МВт', 'АЭС МВт'])
plt.xlabel('Энергообъединение ОЭС')
plt.ylabel('Мощность, МВт')
plt.title('Мощность энергообъединений ОЭС')
plt.show()
```
