**Лабораторная работа №2 Вариант №8 Дан посекундный график освещенности.** Файл “light.csv”. Построить алгоритм, управляющий освещением. По заданию при освещенности примерно ниже 35 условных единиц должен включаться свет, при освещенности примерно выше 35 условных единиц - выключаться.

```python
import csv

#import time #добавил задержку в 1 секунду при выводе значений освещенности, при желании можно включить)

def control_lighting(file_path):
    light_on = False

    with open(file_path, newline='') as csvfile:
        reader = csv.reader(csvfile)
        

        for row in reader:
            light_level = float(row[0])  
            if light_level < 35:
                light_on = True
                print("\033[32mсвет включен.\033[0m")
            elif light_level > 35:
                light_on = False
                print("\033[31mсвет выключен.\033[0m")
            #time.sleep(1)  # Задержка в 1 секунду

control_lighting("light.csv")
```
